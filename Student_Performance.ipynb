```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('StudentPerformanceFactors.csv')
def descriptive_statistics(df):
    """
    Function to get descriptive statistics of the dataset.
    """
    stats = df.describe()
    print("Descriptive Statistics:\n", stats)
    return stats
def plot_correlation_matrix(df):
    """
    Function to plot correlation matrix heatmap to explore relationships between numeric factors.
    """
    numeric_df = df.select_dtypes(include=['float64', 'int64'])
    plt.figure(figsize=(12, 10))
    correlation = numeric_df.corr()
    sns.heatmap(correlation, annot=True, cmap="coolwarm", square=True, fmt=".2f")
    plt.title("Correlation Matrix Heatmap")
    plt.show()
def plot_bar_chart(df, category_column, value_column):
    """
    Function to plot a bar chart for a given category and value.
    """
    plt.figure(figsize=(10, 6))
    sns.barplot(x=category_column, y=value_column, data=df)
    plt.title(f"Bar Chart of {value_column} by {category_column}")
    plt.xlabel(category_column)
    plt.ylabel(value_column)
    plt.xticks(rotation=45)
    plt.show()
def plot_scatter(df, x_column, y_column):
    """
    Function to plot a scatter plot between two columns.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x_column, y=y_column, data=df)
    plt.title(f"Scatter Plot of {y_column} vs {x_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

if __name__ == "__main__":
    stats = descriptive_statistics(data)
    plot_correlation_matrix(data)
    plot_bar_chart(data, 'Hours_Studied', 'Exam_Score')
    plot_scatter(data, 'Attendance', 'Exam_Score')
```
